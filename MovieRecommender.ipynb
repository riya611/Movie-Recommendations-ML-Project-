```python
def fetch_cast(data):
    lst = []
    counter = 0
    for i in ast.literal_eval(data):
        if counter!= 3:
            lst.append(i['name'])
            counter += 1
        else:
            break
    return lst
```

### Using NLP concepts to preprocess the text data

```python
!pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    
    stemmed_text = []
    
    for i in text.split():
        stemmed_text.append(ps.stem(i))
        
    return " ".join(stemmed_text)
```

```python
stem(new_df['tags'][0])
```

### Final function to recommend a movie

```python
distance = similarity[0]
distance
```

```python
def recommend(movie):
    
    movie_index = new_df[new_df['title'] == movie].index[0]    
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    
    
    for i in movie_list:

        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
